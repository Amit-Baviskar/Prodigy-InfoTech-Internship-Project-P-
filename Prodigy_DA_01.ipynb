## **Data Explorating Analysis :**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Transpose the data for easier plotting
df_transposed = df_1960_2022.set_index('Country Name').transpose()

# Calculate the annual sum of the population
annual_sums = df_transposed.sum(axis=1)

# Create the bar plot
plt.figure(figsize=(14, 8))
# Use the Tab10 color palette for unique colors
colors = sns.color_palette("tab10", len(annual_sums))

# Create the bar plot with a different color for each bar
plt.bar(
    annual_sums.index,
    annual_sums.values,
    color=colors
)

# Add labels and title with improved font styling
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Total Population", fontsize=14, fontweight='bold')
plt.title("Total World Population by Year (1960-2022)", fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha="right", fontsize=12)

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# rise in percentage compared to previous year
year=1960
for i in range(len(annual_sums)-1):
  rise=round(((annual_sums[i+1]-annual_sums[i])/annual_sums[i])*100,2)
  print(f"{year+1} has a rise of {rise} in percentage compared to previous year {year}")
  year=year+1
```

## **Insights :**

- **Peak Population Growth (1963-1971) :** The world experienced its highest population growth rate between 1963 and 1971, with a consistent rise above 2% annually. This indicates a period of significant demographic expansion.

- **Declining Growth Rate (Post-1990) :** From the 1990s onward, the population growth rate shows a steady decline, with rises falling below 1.5% after 1996 and gradually approaching less than 1% in recent years.

- **Sharp Decline in Growth (2020-2022) :** A notable drop in the growth rate is observed from 2020 (1.03%) to 2022 (0.8%), potentially attributable to global events such as the pandemic's impact on birth rates and other demographic factors.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data by income levels
income_levels = ['High', 'Upper Middle', 'Lower Middle', 'Low']
filtered_data = df[df['IncomeGroup'].isin(income_levels)]

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)
axes = axes.flatten()

# Loop through each income level and create a plot
for i, income in enumerate(income_levels):
    sns.histplot(
        data=filtered_data[filtered_data['IncomeGroup'] == income],
        x='Total Population',  # Column for population
        kde=True,
        log_scale=True,
        ax=axes[i],
        color=sns.color_palette("coolwarm")[i]  # Using Coolwarm palette for different colors
    )
    axes[i].set_title(f'{income} Income Group', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Total Population (log scale)', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Frequency', fontsize=12, fontweight='bold')

# Adjust layout and add a main title
fig.suptitle('Distribution of Population by Income Levels', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust to leave space for the title
plt.show()
```


## **Insights :**

- **High-Income Group:**  
The population distribution is moderately skewed, with the majority of countries having mid-range population sizes. There is a noticeable lack of countries with exceptionally large populations within this income group.

- **Upper-Middle Income Group:**  
The distribution is fairly uniform, indicating a balanced spread of population sizes among countries at this income level. However, a handful of nations account for a significant portion of the higher population values.

- **Lower-Middle Income Group:**  
The distribution shows a slight right skew, signifying that while many countries have mid-to-low population figures, there are a few with notably large populations. This income group plays a major role in the global population count.

- **Low-Income Group:**  
The population distribution is predominantly centered around smaller population figures, with the majority of countries having relatively low population sizes in comparison to higher-income groups.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data by the specified regions
regions = [
    'Latin America & Caribbean', 'South Asia', 'Sub-Saharan Africa',
    'Europe & Central Asia', 'Middle East & North Africa',
    'East Asia & Pacific', 'North America'
]
filtered_data = df[df['Region'].isin(regions)]

# Create subplots
fig, axes = plt.subplots(4, 2, figsize=(16, 15), sharex=True, sharey=True)
axes = axes.flatten()

# Loop through each region and create a plot
for i, region in enumerate(regions):
    if i < len(axes):  # Ensure we don't exceed subplot slots
        sns.histplot(
            data=filtered_data[filtered_data['Region'] == region],
            x='Total Population',  # Column for population
            kde=True,
            log_scale=True,
            ax=axes[i],
            color=sns.color_palette("Set2")[i]  # Changed to Set2 color palette
        )
        axes[i].set_title(f'{region}', fontsize=14, fontweight='bold')
        axes[i].set_xlabel('Total Population (log scale)', fontsize=12, fontweight='bold')
        axes[i].set_ylabel('Frequency', fontsize=12, fontweight='bold')

# Hide any unused axes (if regions < axes slots)
for j in range(len(regions), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and add a main title
fig.suptitle('Distribution of Population by Region', fontsize=18, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust to leave space for the title
plt.show()
```

### **Conclusion:**

The analysis of global population trends highlights important demographic shifts influenced by regional, economic, and societal factors. Key observations include:

1. **Shift from High Growth to Stabilization**  
   - Global population growth peaked in the mid-20th century, but more recently, growth rates have slowed significantly, dropping below 1% after 2020. This marks a shift towards demographic equilibrium in many parts of the world, driven by lower fertility rates, urbanization, and changing socio-economic conditions.

2. **Regional Differences**  
   - Regions like **South Asia** and **Sub-Saharan Africa**, with rapid population growth, offer considerable prospects for economic development. However, they also face challenges related to infrastructure, resource management, and social inequality.

3. **Economic Growth Potential**  
   - **Lower- and upper-middle-income countries** are home to the majority of the global population, representing significant opportunities for economic expansion. Focused investments in education, healthcare, and job creation could spur further growth and development in these regions.

4. **Challenges and Prospects for the Future**  
   - Emerging economies, especially in Africa and Asia, must leverage their youthful populations by implementing policies aimed at developing a skilled workforce, which can drive future growth.
   - At the same time, aging populations in wealthier regions present challenges for workforce sustainability and social support systems, necessitating forward-thinking solutions to maintain economic productivity.
